```python
import ropes_league
```

```python
data = ropes_league.LeagueData.from_env()
```

```python
scores = data.season_scores().assign(
    top_diff=lambda df: df[["difficulty_1", "difficulty_2", "difficulty_3"]].max(axis=1),
    bot_diff=lambda df: df[["difficulty_1", "difficulty_2", "difficulty_3"]].min(axis=1),
    n_climbs=lambda df: df[["difficulty_1", "difficulty_2", "difficulty_3"]].count(axis=1)
)
```

```python
scores
```

```python
climbers = data.climbers().assign(
    points=scores.groupby("climber_id").points.sum(),
    nights=scores.groupby("climber_id").points.count(),
    top_diff=scores.groupby("climber_id").top_diff.max(),
    bot_diff=scores.groupby("climber_id").bot_diff.min(),
    top_grade=lambda df: df.top_diff.replace(
        {v: k for k, v in ropes_league.league_data.GRADE_DIFFICULTY.items()}
    ),
    bot_grade=lambda df: df.bot_diff.replace(
        {v: k for k, v in ropes_league.league_data.GRADE_DIFFICULTY.items()}
    ),
    diff_diff=lambda df: df.top_diff - df.bot_diff,
    best_week_score=scores.groupby("climber_id").points.max(),
    best_week=scores.loc[scores.groupby("climber_id").points.idxmax()].week.reset_index(
        drop=True
    ),
    n_climbs=scores.groupby("climber_id").n_climbs.sum(),
)
```
