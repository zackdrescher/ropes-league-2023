```python
climbs = climbs.assign(
    avg_diff=climbs[["difficulty_1", "difficulty_2", "difficulty_3"]].mean(axis=1)
)
```

```python
climbs.drop('name', axis=1).merge(
    climbers[["name", "difficulty", "grade"]].rename(
        columns={"difficulty": "climber_difficulty"}
    ),
    left_on="climber_id",
    right_index=True,
).assign(diff_delta=lambda df: df.avg_diff - df.climber_difficulty)[
    [
        "name",
        "grade",
        "grade_1",
        "grade_2",
        "grade_3",
        "avg_diff",
        "diff_delta",
    ]
].sort_values(
    "diff_delta", ascending=False
)
```
