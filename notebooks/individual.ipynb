```python
from ropes_league.league_data import climb_string


def week_results_string(scores_data):
    return "\n".join(
        [
            f"{i+1}. {row['name']} - {climb_string(row)} - {row['score']}"
            for i, row in scores_data.reset_index().iterrows()
        ]
    )
```
